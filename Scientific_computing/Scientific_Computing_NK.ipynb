**probability-calculator**

```python
import copy
import random
# Consider using the modules imported above.

class Hat:
    def __init__(self, **contents):
        self.contents = list()
        for key, value in contents.items():
            x=0
            while x < value:
                self.contents.append(key)
                x = x+1

    def draw(self, num_draws):
        if num_draws > len(self.contents):
            num_draws = len(self.contents)
        draws = random.sample(self.contents, k=num_draws)
        for draw in draws:
            self.contents.remove(draw)
        return draws

    def __str__(self):
        return str(self.contents)

def experiment(hat, expected_balls, num_balls_drawn, num_experiments):
    M = 0
    x = 0
    while x < num_experiments:
        hat_copy = copy.deepcopy(hat)
        expected_balls_list = list()
        for key, value in expected_balls.items():
            y = 0
            while y < value:
                expected_balls_list.append(key)
                y = y + 1

        draws = hat_copy.draw(num_balls_drawn)
        for draw in draws:
            try:
                expected_balls_list.remove(draw)
            except:
                continue

        if len(expected_balls_list) == 0:
            M = M + 1

        x = x + 1

    return M/num_experiments


#prob_calculator.random.seed(95)
hat = Hat(blue=9, red=2, green=6)
probability = experiment(
    hat=hat,
    expected_balls={"blue": 2,
                    "red": 1},
    num_balls_drawn=4,
    num_experiments=3000)
print("Probability:", probability)
```

**Shape-Calculator**

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.area = 0
        self.perimeter = 0
        self.diagonal = 0

    def set_width(self, value):
        self.width = value

    def set_height(self, value):
        self.height = value

    def get_area(self):
        self.area = self.width * self.height
        return self.area

    def get_perimeter(self):
        self.perimeter = 2 * self.width + 2 * self.height
        return self.perimeter

    def get_diagonal(self):
        self.diagonal = (self.width ** 2 + self.height ** 2) ** 0.5
        return self.diagonal

    def get_picture(self):
        if self.width > 50 or self.height > 50:
            return "Too big for picture."
        else:
            pic = ""
            x = 1
            while x <= self.height:
                pic = pic + self.width * "*" + "\n"
                x = x + 1
            return pic

    def get_amount_inside(self, shape):
        times_width = int(self.width / shape.width)
        times_height = int(self.height / shape.height)
        inside = times_width * times_height
        return inside

    def __str__(self):
        return "Rectangle(width=" + str(self.width) + ", height=" + str(self.height) + ")"


class Square(Rectangle):

    def __init__(self, side_length, width=0, height=0):
        super().__init__(width, height)
        self.width = side_length
        self.height = side_length

    def set_side(self, value):
        self.width = value
        self.height = value

    def set_width(self, value):
        self.width = value
        self.height = value

    def set_height(self, value):
        self.width = value
        self.height = value

    def __str__(self):
        return "Square(side=" + str(self.width) + ")"


rect = Rectangle(5, 10)
print(rect.get_area())
rect.set_width(3)
print(rect.get_perimeter())
print(rect)

sq = Square(9)
print(sq.get_area())
sq.set_side(4)
print(sq.get_diagonal())
print(sq)
```

**budget-app**

```python
class Category:

    def __init__(self, name):  # giving name as parameter to an instance
        self.name = name
        self.ledger = list()  # giving many instance attributes
        self.balance = 0
        self.spent = 0
        self.percent = 0
        self.hundret = "   "
        self.ninty = "   "
        self.eighty = "   "
        self.seventy = "   "
        self.sixty = "   "
        self.fifty = "   "
        self.forty = "   "
        self.thirty = "   "
        self.twenty = "   "
        self.ten = "   "
        self.zero = "   "

    def deposit(self, amount, description=""):  # defining the deposit methode which takes amount and description as parameter
        self.amount = amount
        self.descr = description
        self.ledger.append({"amount": amount, "description": description})  # each list entry is a dict
        self.balance = self.balance + self.amount

    def withdraw(self, amount, description=""):
        self.amount = amount
        self.descr = description

        if self.balance - self.amount >= 0:
            self.balance = self.balance - self.amount
            self.amount = self.amount * -1
            amount = self.amount
            self.ledger.append({"amount": amount, "description": description})
            self.spent = self.spent + self.amount
            return True
        else:
            return False

    def get_balance(self):
        return self.balance

    def transfer(self, amount, categ):  # method with amount and categ parameter
        self.amount = amount

        if self.amount <= self.balance:
            self.withdraw(self.amount, "Transfer to " + categ.name)  # categ.name to define name for transfer target
            categ.deposit(self.amount * -1, "Transfer from " + self.name)
            return True
        else:
            return False

    def check_funds(self, amount):
        if amount > self.balance:
            return False
        else:
            return True

    def __str__(self):  # method to define what will be printed out for the instance
        title = '{:*^{width}}'.format(self.name, width=30)
        items = ""
        for entry in self.ledger:  # each list entry is a dict, turn dict into string
            value = list(entry.values())
            if len(value[1]) <= 23:
                descr = '{:<{width}}'.format(value[1], width=23)
            else:
                descr = value[1][:23]
            amount = value[0]
            amount = '{:>7.2f}'.format(float(amount))

            items = items + descr + amount + "\n"

        return title + "\n" + items + "Total: " + str(self.balance)


def create_spend_chart(categories):

    title = "Percentage spent by category"
    sum_spent = 0
    for category in categories:  # total spending
        sum_spent = sum_spent + category.spent

    for category in categories:  # calculates % spent per category
        x = int(category.spent * 100 / sum_spent)
        if x >= 10:  # round function does not round down single digits to 0, if statement is a workaround
            category.percent = round(int(category.spent * 100 / sum_spent), -1)
        else:
            category.percent = 0

    for category in categories:
        below = False
        if category.percent == 100:
            below = True
            category.hundret = " o "
        if category.percent == 90 or below is True:
            below = True
            category.ninty = " o "
        if category.percent == 80 or below is True:
            below = True
            category.eighty = " o "
        if category.percent == 70 or below is True:
            below = True
            category.seventy = " o "
        if category.percent == 60 or below is True:
            below = True
            category.sixty = " o "
        if category.percent == 50 or below is True:
            below = True
            category.fifty = " o "
        if category.percent == 40 or below is True:
            below = True
            category.forty = " o "
        if category.percent == 30 or below is True:
            below = True
            category.thirty = " o "
        if category.percent == 20 or below is True:
            below = True
            category.twenty = " o "
        if category.percent == 10 or below is True:
            below = True
            category.ten = " o "
        if category.percent == 0 or below is True:
            category.zero = " o "

    col_100 = "100|"
    col_90 = " 90|"
    col_80 = " 80|"
    col_70 = " 70|"
    col_60 = " 60|"
    col_50 = " 50|"
    col_40 = " 40|"
    col_30 = " 30|"
    col_20 = " 20|"
    col_10 = " 10|"
    col_0 = "  0|"

    for category in categories:
        col_100 = col_100 + category.hundret
        col_90 = col_90 + category.ninty
        col_80 = col_80 + category.eighty
        col_70 = col_70 + category.seventy
        col_60 = col_60 + category.sixty
        col_50 = col_50 + category.fifty
        col_40 = col_40 + category.forty
        col_30 = col_30 + category.thirty
        col_20 = col_20 + category.twenty
        col_10 = col_10 + category.ten
        col_0 = col_0 + category.zero

    bar = "    " + (len(col_0) - 3) * "-"

    longest_name = 0
    for category in categories:
        if len(category.name) > longest_name:
            longest_name = len(category.name)
    x = 0
    names = "    "
    while x <= longest_name:
        for category in categories:
            if x < len(category.name):
                names = names + " " + category.name[x] + " "
            else:
                names = names + "   "
        names = names + " \n    "
        x = x + 1
    names = names[:len(names)-20]  # cleaning up last row

    chart = title + "\n" + col_100 + " \n" \
            + col_90 + " \n" \
            + col_80 + " \n" \
            + col_70 + " \n" \
            + col_60 + " \n" \
            + col_50 + " \n" \
            + col_40 + " \n" \
            + col_30 + " \n" \
            + col_20 + " \n" \
            + col_10 + " \n" \
            + col_0 + " \n" + bar + "\n" + names
    return chart


food = Category("Food")
food.deposit(1000, "initial deposit")
food.withdraw(10.15, "groceries")
food.withdraw(15.89, "restaurant and more food for dessert")
print(food.get_balance())
clothing = Category("Clothing")
food.transfer(50, clothing)
clothing.withdraw(25.55)
clothing.withdraw(100)
auto = Category("Auto")
auto.deposit(1000, "initial deposit")
auto.withdraw(15)

print(food)
print(clothing)

print(create_spend_chart([food, clothing, auto]))
```

**time calculator**

```python
def add_time(start, duration, day=""):
    start = start.split()
    time = start[0]
    time_hr, time_min = time.split(":")
    am_pm = start[1]
    duration_hr, duration_min = duration.split(":")

    # compute new minute value and sum additional hours
    sum_min = int(time_min) + int(duration_min)
    if sum_min > 60:
        min_to_hr = sum_min/60
        new_min = sum_min - int(min_to_hr)*60
        add_hr = int(duration_hr) + int(min_to_hr)
    else:
        new_min = sum_min
        add_hr = int(duration_hr)

    # compute new hour
    sum_hr = int(time_hr) + add_hr - int(add_hr/12) * 12  # removing multiples of 12 from add_hr
    if sum_hr > 12:
        new_hr = sum_hr - 12
    else:
        new_hr = sum_hr

    # determine AM/PM; if PM to AM PM_to_AM = True
    PM_to_AM = False
    half_days = (int(time_hr) + add_hr)/12
    if int(half_days) % 2 != 0:  # if odd number of half days (multiples of 12)
        if am_pm == "AM":
            am_pm = "PM"
        else:
            am_pm = "AM"
            PM_to_AM = True

    # add next day info
    day_info = ""
    if PM_to_AM is True and int(half_days) == 1:
        day_info = " (next day)"
    elif int(half_days) == 2:
        day_info = " (next day)"
    elif int(half_days) == 3 and PM_to_AM is False:
        day_info = " (next day)"
    elif int(half_days) > 2 and int(half_days) % 2 != 0:        # if odd num of half days
        if PM_to_AM is True:
            day_info = " (" + str(int(half_days/2) + 1) + " days later)"
        else:
            day_info = " (" + str(int(half_days / 2)) + " days later)"
    elif int(half_days) > 2 and int(half_days) % 2 == 0:        # if even num of half days
        day_info = " (" + str(int(half_days/2)) + " days later)"

    # determine new day name
    new_day = day
    if day != "":
        all_days = {
            0: "Monday",
            1: "Tuesday",
            2: "Wednesday",
            3: "Thursday",
            4: "Friday",
            5: "Saturday",
            6: "Sunday"
        }
        all_days_val = list(all_days.values())
        day_num = all_days_val.index(day.capitalize())
        day_sum = day_num + int(half_days / 2) - int(int(half_days / 2) / 7) * 7  # removing multiples of 7
        if day_sum > 6:
            new_day_num = day_sum - 7
        else:
            new_day_num = day_sum
        if PM_to_AM is True:
            new_day_num = new_day_num + 1
            if new_day_num > 6:
                new_day_num = new_day_num - 7
        new_day = ", " + all_days[new_day_num]

    # return new date
    new_hr_min = str(new_hr) + ":" + str(new_min).rjust(2, '0')
    end = new_hr_min + " " + am_pm
    new_time = end + new_day + day_info

    return new_time

print(add_time("11:06 PM", "2:02", 'Monday'))
```

**arithmetic formatter**

```python
def arithmetic_arranger(inputs, give_results=False):
    # convert input into 2D list
    all_problems = list()
    for input in inputs:
        math = input.split()
        all_problems.append(math)

    # check error cases
    if len(all_problems) > 5:
        problem_return = "Error: Too many problems."
        return problem_return

    for problem in all_problems:
        if not problem[0].isdigit() or not problem[2].isdigit():
            problem_return = "Error: Numbers must only contain digits."
            return problem_return

        elif problem[1] != "+" and problem[1] != "-":
            problem_return = "Error: Operator must be '+' or '-'."
            return problem_return

        elif len(problem[0]) > 4 or len(problem[2]) > 4:
            problem_return = "Error: Numbers cannot be more than four digits."
            return problem_return

        else:
            continue

    # formate line 1,2,3
    sl1 = list()
    sl2 = list()
    i = 0
    for problem in all_problems:
        space1 = max(len(all_problems[i][0]), len(all_problems[i][2])) + 2 - len(all_problems[i][0])
        sl1.append(space1)
        space2 = max(len(all_problems[i][0]), len(all_problems[i][2])) + 1 - len(all_problems[i][2])
        sl2.append(space2)
        i = i + 1

    i = 0
    l1 = ""
    for problem in all_problems:
        handle = (sl1[i] * " " + all_problems[i][0] + "    ")
        l1 = l1 + handle
        i = i + 1
    l1 = l1[:len(l1)-4]

    i = 0
    l2 = ""
    for problem in all_problems:
        handle = (all_problems[i][1] + sl2[i] * " " + all_problems[i][2] + "    ")
        l2 = l2 + handle
        i = i + 1
    l2 = l2[:len(l2) - 4]

    i = 0
    l3 = ""
    for problem in all_problems:
        handle = ((max(len(all_problems[i][0]), len(all_problems[i][2])) + 2) * "-" + "    ")
        l3 = l3 + handle
        i = i + 1
    l3 = l3[:len(l3) - 4]
    problem_return = l1 + "\n" + l2 + "\n" + l3

    # compute results and formate output line 4
    if give_results == True:
        results = list()
        for problem in all_problems:
            if problem[1] == "+":
                result = int(problem[0]) + int(problem[2])
            else:
                result = int(problem[0]) - int(problem[2])
            results.append(result)

        i = 0
        l4 = ""
        for result in results:
            handle = (((max(len(all_problems[i][0]), len(all_problems[i][2])) + 2) - len(str(result))) * " " + str(
                result) + "    ")
            l4 = l4 + handle
            i = i + 1
        l4 = l4[:len(l4) - 4]
        problem_return = l1 + "\n" + l2 + "\n" + l3 + "\n" + l4
    return problem_return


print(arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"], True))
```
