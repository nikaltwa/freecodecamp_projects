```python
import numpy as np

def calculate(list):
    if len(list) != 9:
        raise ValueError('List must contain nine numbers.')
        #calculations = "List must contain nine numbers."
        #return calculations

    list_arr = [[list[0], list[1], list[2]],
                [list[3], list[4], list[5]],
                [list[6], list[7], list[8]]]
    arr = np.array(list_arr)
    #print(arr)

    mean_axis1 = arr.mean(0)
    mean_axis2 = arr.mean(1)
    mean_flattened = arr.flatten().mean()
    mean_axis1 = mean_axis1.tolist()
    mean_axis2 = mean_axis2.tolist()
    mean_flattened = mean_flattened.tolist()

    variance_axis1 = arr.var(0)
    variance_axis2 = arr.var(1)
    variance_flattened = arr.flatten().var()
    variance_axis1 = variance_axis1.tolist()
    variance_axis2 = variance_axis2.tolist()
    variance_flattened = variance_flattened.tolist()

    std_dev_axis1 = arr.std(0)
    std_dev_axis2 = arr.std(1)
    std_dev_flattened = arr.flatten().std()
    std_dev_axis1 = std_dev_axis1.tolist()
    std_dev_axis2 = std_dev_axis2.tolist()
    std_dev_flattened = std_dev_flattened.tolist()

    max_axis1 = arr.max(0)
    max_axis2 = arr.max(1)
    max_flattened = arr.flatten().max()
    max_axis1 = max_axis1.tolist()
    max_axis2 = max_axis2.tolist()
    max_flattened = max_flattened.tolist()

    min_axis1 = arr.min(0)
    min_axis2 = arr.min(1)
    min_flattened = arr.flatten().min()
    min_axis1 = min_axis1.tolist()
    min_axis2 = min_axis2.tolist()
    min_flattened = min_flattened.tolist()

    sum_axis1 = arr.sum(0)
    sum_axis2 = arr.sum(1)
    sum_flattened = arr.flatten().sum()
    sum_axis1 = sum_axis1.tolist()
    sum_axis2 = sum_axis2.tolist()
    sum_flattened = sum_flattened.tolist()

    calculations = {  # Key: Value >> Key must be unique
        "mean": [mean_axis1, mean_axis2, mean_flattened],  # axis1 = vertical, axis2 = horizontal, flattened=all values
        "variance": [variance_axis1, variance_axis2, variance_flattened],
        "standard deviation": [std_dev_axis1, std_dev_axis2, std_dev_flattened],
        "max": [max_axis1, max_axis2, max_flattened],
        "min": [min_axis1, min_axis2, min_flattened],
        "sum": [sum_axis1, sum_axis2, sum_flattened]
    }

    return calculations

print(calculate([0,1,2,3,4,5,6,7,8]))
```
